```python
import numpy as np

# Create one-hot encoding vectors
def one_hot_encoding(index, size):
    vec = np.zeros(size, dtype=int)
    vec[index] = 1
    return vec.tolist()
```

```python
import hashlib

# Function to generate binary hash code
def get_binary_hash(item, num_bits):
    # Generate a hash (e.g., SHA-256)
    hash_object = hashlib.sha256(item.encode())
    # Convert hash to binary and truncate to the desired number of bits
    binary_hash = bin(int(hash_object.hexdigest(), 16))[2:].zfill(256)[:num_bits]
    return binary_hash
```

```python
# Genotype encoding
gts = depmap_mutation_ref["gt"].unique().tolist()
encoded_gts = {key: one_hot_encoding(i, len(keys)) for i, key in enumerate(gts)}
```

```python
# Mutation type encoding
variant_types = depmap_mutation_ref["variant_type"].unique().tolist()
encoded_variant_types = {key: one_hot_encoding(i, len(keys)) for i, key in enumerate(variant_types)}
```

```python
# Chromosome encoding
chroms = depmap_mutation_ref["chrom"].unique().tolist()
num_bits = 6
encoded_chroms = {item: get_binary_hash(item, num_bits) for item in chroms}
for item, binary_hash in encoded_chroms.items():
    print(f"{item}: {binary_hash}")
```

```python
# Gene encoding
genes = depmap_mutation_ref["gene"].unique().tolist()
num_bits = 16 
encoded_genes = {item: get_binary_hash(item, num_bits) for item in genes}

count = 0
for item, binary_hash in encoded_genes.items():
    print(f"{item}: {binary_hash}")
    count += 1
    if count > 10:
        break
```

```python
import math

# Position encoding
def get_position_encoding(pos, dim):
    encoding = []
    for i in range(dim):
        index = i+1
        if (index%2 == 0):
            encoding.append(math.sin(pos/math.pow(10000, 2*index/dim)))
        else:
            encoding.append(math.cos(pos/math.pow(10000, 2*index/dim)))
    return encoding
```

### Alternative: each mutation has 36 dimension (like one word in a document)
- chro: 6d: binary, hashing
- pos: 4d: real value, wave
- af: 1d: real value, self
- gt: 4d: binary, one-hot
- variant_type: 4d: binary, one-hot
- hotspot: 1d: binary, self
- gene: 16d: binary, hashing

```python
from rdkit import Chem
from rdkit.Chem.rdFingerprintGenerator import GetMorganGenerator  # Correct import for new RDKit
import numpy as np
import pubchempy as pcp

def get_smiles(drug_name):
    compounds = pcp.get_compounds(drug_name, 'name')
    if compounds:
        return compounds[0].isomeric_smiles  # Returns the first match's SMILES
    return None  # Return None if no results found

def get_morgan_fingerprint(smiles, radius=2, n_bits=2048):
    mol = Chem.MolFromSmiles(smiles)  # Convert SMILES to molecule
    if mol is None:
        return None  # Invalid SMILES
    
    generator = GetMorganGenerator(radius=radius, fpSize=n_bits)  # New recommended method
    fingerprint = generator.GetFingerprint(mol)

    # Convert to NumPy array
    fingerprint_bits = np.array(fingerprint, dtype=int)

    return fingerprint_bits
```
